```python
%%writefile test.hql

use stackoverflow_;

DROP VIEW IF EXISTS ages_qa;

CREATE VIEW ages_qa
AS
SELECT users.age,
       SUM(qa.qcount),
       SUM(qa.acount)
FROM 
(SELECT  owner_user_id as uid,
        COUNT(IF(post_type_id=1, 1, null)) as qcount,
        COUNT(IF(post_type_id=2, 1, null)) as acount
FROM posts
WHERE owner_user_id is not null
GROUP BY owner_user_id) qa
JOIN users
ON users.id = qa.uid
WHERE users.age is not null
GROUP BY users.age
SORT BY users.age;

SELECT *
 FROM ages_qa;
```
