```python
%%writefile test.hql

use stackoverflow_;

DROP VIEW IF EXISTS toptags2016;

CREATE VIEW toptags2016
AS 
SELECT y16.tcount16,
       y16.tag,
       RANK() OVER (ORDER BY y16.tcount16 DESC) as rank16
FROM
(SELECT COUNT(1) as tcount16,
       tag
 FROM posts
 LATERAL VIEW explode(tags) tagslist AS tag
 WHERE post_type_id=1 AND year='2016'
 GROUP BY tag) y16
LIMIT 10;

DROP VIEW IF EXISTS tags2009;

CREATE VIEW tags2009
AS 
SELECT y09.tcount09,
       y09.tag,
       RANK() OVER (ORDER BY y09.tcount09 DESC) as rank09
FROM
(SELECT COUNT(1) as tcount09,
       tag
 FROM posts
 LATERAL VIEW explode(tags) tagslist AS tag
 WHERE post_type_id=1 AND year='2009'
 GROUP BY tag) y09;

DROP VIEW IF EXISTS compare16vs09;

CREATE VIEW compare16vs09
AS
SELECT toptags2016.tag,
       toptags2016.rank16 as r,
       tags2009.rank09,
       toptags2016.tcount16,
       tags2009.tcount09
FROM toptags2016
LEFT JOIN tags2009
ON toptags2016.tag = tags2009.tag
SORT BY r;

SELECT *
 FROM compare16vs09;
```
